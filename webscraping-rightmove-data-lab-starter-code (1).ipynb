```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
import time
import numpy as np
```

```python
url = 'https://www.rightmove.co.uk/property-for-sale/find.html?locationIdentifier=REGION%5E87490&index=0'
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup.text
```

```python
print(soup.prettify)
```

```python
for s in soup.find_all('h2'):
    print(s.text)
```

```python
location = []
t=soup.find_all('address', class_="propertyCard-address")
for i in t:
    location.append(i.text.strip())
print(location)
```

```python
desc = []
d = soup.find_all("span", itemprop = "description")
for i in d:
    desc.append(i.text.strip())
print(desc)
```

```python
price = []
p = soup.find_all("div", class_ = "propertyCard-priceValue")
for i in p:
    price.append(i.text.strip())
print(price)
```

```python
df = pd.DataFrame(list(zip(location, desc, price)), columns = ['Location', 'Description', 'Price'])
```

```python
location = []
desc = []
price = []   

for x in range(43):
    url = f'https://www.rightmove.co.uk/property-for-sale/find.html?locationIdentifier=REGION%5E87490&index={x}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    t=soup.find_all('address', class_="propertyCard-address")
    for i in t:
        location.append(i.text.strip())
    d = soup.find_all("span", itemprop = "description")
    for i in d:
        desc.append(i.text.strip())
    p = soup.find_all("div", class_ = "propertyCard-priceValue")
    for i in p:
        price.append(i.text.strip())


df = pd.DataFrame(list(zip(location, desc, price)), columns = ['Location', 'Description', 'Price'])    

```
